```python
#importa a tabela com os dados de todos os municípios
mun = pd.read_csv('municipio.csv')
mun.head()
```

```python
#importa a tabela com os dados do PIB por município
pib = pd.read_csv('pibMunicipio.csv')
pib.head()
```

```python
#unifica as tabelas de município com seus respectivos PIBs
#sintaxe pd.merge(tabela_da_esquerda, tabela_da_direita, on="coluna_coincidente", how="left|right|inner|outer)
grl = pd.merge(mun, pib, on='id_municipio', how='left')
grl.head()
```
